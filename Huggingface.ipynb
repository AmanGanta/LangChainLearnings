```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
import numpy as np
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
```

```python
l=PyPDFLoader(r"C:\Users\amang\Downloads\proj celeb search by krish naik\1706.03762v7.pdf")
dc=l.load()
tex_spl=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
fdc=tex_spl.split_documents(dc)
```

```python
fdc
```

```python
model_name = "BAAI/bge-small-en"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
hf = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)
```

```python
hf.embed_query(fdc[0].page_content)
```

```python
vs=FAISS.from_documents(fdc,hf)
```

```python
rt=vs.as_retriever(search_type="similarity",search_kwargs={"k":2})
print(rt)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ['HUGGINGFACEHUB_API_TOKEN']=os.getenv("HUGGINGFACEHUB_API_TOKEN")
```

```python
from langchain_community.llms import HuggingFaceHub
lm=HuggingFaceHub(
    repo_id="mistralai/Mistral-7B-v0.1",
    model_kwargs={"temperature":0.1,"max_length":500}
)
query="dimention of encoder output"
lm.invoke(query)
```

```python
from langchain_community.llms.huggingface_pipeline import HuggingFacePipeline
hf = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    pipeline_kwargs={"max_new_tokens": 300}
)
llm = hf
llm.invoke(query)
```

```python
prompt_template="""
Use the following piece of context to answer the question asked.
Please try to provide the answer only based on the context

{context}
Question:{question}

Helpful Answers:
 """
```

```python
p=PromptTemplate(template=prompt_template,input_variables=["context","question"])
```

```python
retQA=RetrievalQA.from_chain_type(
    llm=hf,
    chain_type="stuff",
    retriever=rt,
    return_source_documents=True,
    chain_type_kwargs={"prompt":p}
)
```

```python
q="dimention of encoder output"
r = retQA.invoke({"query": query})
print(r['result'])
```
